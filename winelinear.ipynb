```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
whiteData = pd.read_csv('data/winequality-white.csv',sep=';')

redData = pd.read_csv('data/winequality-red.csv',sep=';')
```

```python
whiteData.corr()
```

```python
redData.corr()
```

```python
#We can See that for White data, Alcohol, with .435 and Density with -.3 has the best correlation 
```

```python
#For red, we have alcohol with .476, and volatile acidity with -.39
```

```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

whiteModel = LinearRegression()
redModel = LinearRegression()
```

```python
x = whiteData['alcohol'].to_numpy()
x = x.reshape(-1, 1)
y = whiteData['quality']
whiteModel.fit(x,y)

X2 = sm.add_constant(x)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```

```python
r2_score = whiteModel.score(x, y)
print(f"R-squared value: {r2_score}")
```
